```python
from gensim.models import Word2Vec
from gensim.models import KeyedVectors
from gensim.test.utils import datapath
from gensim.models.phrases import Phrases, Phraser
import pandas as pd
from nltk.tokenize import word_tokenize
import os
```

```python
# CSV 파일 경로
folder_path = r'C:\Users\SDA04\Desktop\yanghs81\proj_프로젝트\2_final_ESG\token_folder'

#sentences를 저장할 딕셔너리
sentences = []
i = 0
for file_name in os.listdir(folder_path):
    if file_name.endswith('.csv'):
        print(file_name)
        file_path = os.path.join(folder_path, file_name)
        df = pd.read_csv(file_path)
        

        for index, row in df.iterrows():
            tokens = row['token']
            temp = eval(tokens)
            sentences.append(temp)
            
model = Word2Vec(sentences, vector_size=100, window=5, min_count=1, workers=4)
model.save('word2vec.model')
```

```python
# 주어진 단어와 유사한 단어 찾기
s_list = ['환경','녹색','신재생','에너지','오염','일자리','사회','정보','산업','안전','장애인','육아','지역경제','부패','준법','권리','소송','공정','세금','탈세','배임','횡령','감사','경영','재정']
similar_words = []
for word in s_list:
    similar_words.append(model.wv.most_similar(word, topn=10))
```

```python
# 주어진 단어와 유사한 단어를 찾고 주어진 단어도 넣기
result = []
s_list = ['환경','녹색','신재생','에너지','오염','일자리','사회','정보','산업','안전','장애인','육아','지역경제','부패','준법','권리','소송','공정','세금','탈세','배임','횡령','감사','경영','재정']
i = 0
for word in similar_words:
    k_list = []
    k_list.append(s_list[i])
    i += 1
    for a,b in word:
        k_list.append(a)
    result.append(k_list)
    print(k_list)
print(result)
```

```python
# 환경과 관련된 단어들
environment_related_words = ["environment", "climate", "sustainability", "ecosystem", "conservation"]

# 각 기사의 단어 리스트 생성
article_words = [word for sentence in sentences for word in sentence]

# 환경과 관련된 단어가 포함된 기사 찾기
environment_related_articles = []
for sentence in sentences:
    if any(word in sentence for word in environment_related_words):
        environment_related_articles.append(" ".join(sentence))

# 찾아진 기사 출력
for article in environment_related_articles:
    print(article)
    print("-------------")
```
